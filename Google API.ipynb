## Connecting to Mongo to store the coords in collections

```python
#Mongo DB collections parameters 

# auxiliar collection for storing coordinates from kindergartens in New York
kinder_db, kinder_coll = connectCollection('New_York','Kindergarten')

#auxiliar collection for storing coordinates from vegan sites in New York
vegan_db, vegan_coll = connectCollection('New_York','Vegans')

# final collection
final_db, final_coll = connectCollection('New_York','Final_New_York')
```

## Starbucks in New York

```python
#opening the JSON file with the data

with open('./output/Starbucks.json', "r") as read_file:
    starbucks_ny = json.load(read_file)
```

```python
# map showing the starbucks located in New York
map_city = folium.Map(location=[40.730610,-73.935242], zoom_start=12)

for i in range(len(starbucks_ny)):
    folium.Marker([starbucks_ny[i]['lat'],starbucks_ny[i]['lng']],
                    radius=2,
                   
                    icon=folium.Icon(prefix='fa',icon='coffee',color='green'), 
                   ).add_to(map_city)
    

display(map_city)   
```

## Kindergartens in New York

```python
#opening the JSON file with the data

with open('./output/Kindergartens.json', "r") as read_file:
    kindergartens_ny = json.load(read_file)
```

```python
# Some kindergartens are located out of the area
map_city = folium.Map(location=[40.730610,-73.935242], zoom_start=10)

for i in range(len(kindergartens_ny)):
    folium.Marker([kindergartens_ny[i]['lat'],kindergartens_ny[i]['lng']],
                    radius=2,
                   
                    icon=folium.Icon(prefix='fa',icon='child',color='blue'), 
                   ).add_to(map_city)
display(map_city)
```

```python
# Geoquery for selecting the offices that are in New York using a polygon 

kindergartens_ny = list(kinder_coll.find({'coordinates': {'$geoWithin': { '$geometry': 
            { 'type': 'Polygon', 'coordinates': [ [ [ -73.96999075019157, 40.80715105959315 ],
            [ -73.93633235156877, 40.79000330835164 ], [ -73.98407232512561, 40.70263699912775 ],
            [ -74.02288251945599, 40.69665237044275 ], [ -74.02528669078619, 40.72604994282311 ], 
            [ -73.98063779465389, 40.80715105959315 ], [ -73.96999075019157, 40.80715105959315 ]] ] }}}}))
```

```python
# map showing only the kindergartens which are in the area

map_city = folium.Map(location=[40.730610,-73.935242], zoom_start=10)

for i in range(len(kindergartens_ny)):
    folium.Marker([kindergartens_ny[i]['coordinates'][-1],kindergartens_ny[i]['coordinates'][0]],
                    radius=2,
                   
                    icon=folium.Icon(prefix='fa',icon='child',color='blue'), 
                   ).add_to(map_city)
display(map_city)
```
